```python
import matplotlib.pyplot as plt
import math
```

```python
def one_d_tent_map(mu: float, x: float):
    if x < 0.5:
        return mu * x
    else:
        return mu * (1.0 - x)

def plot_tent(seq):

    plt.plot(seq)
    plt.xlabel('i')
    plt.ylabel('x_i')
    plt.title('One Dimensional Tent Map')
    plt.show()
```

```python
x = 0.1
mu = 1.99999
seq = [x]
iter = 2**8

for i in range(iter):
    x = one_d_tent_map(mu, x)
    seq.append(x)

plot_tent(seq)
```

```python
x = 0.5
alpha = 10**-3
seq = [x]
iter = 2**10
lambdas = [1, 1, 1, 1, 1, 1, 10**-2, 0.5]

def multi_d_tent_map(alpha: float, x: float):
    if x < 0.7:
        return (10 / 7) * (1 - alpha * abs(math.cos(lambdas[0])) - alpha * abs(math.sin(lambdas[1]) - alpha * abs(math.atan(lambdas[2])))) * x
    else:
        return (10 / 3) * (1 - alpha * abs(math.cos(lambdas[3])) - alpha * abs(math.sin(lambdas[4]) - alpha * abs(math.atan(lambdas[5])))) * (1 - x)
    
for i in range(iter):
    x = multi_d_tent_map(alpha, x)
    seq.append(x)

plot_tent(seq)
```
